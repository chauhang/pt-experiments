```python
import pandas as pd

df = pd.read_csv('pt_question_answers_updated.csv')
df

import re
CLEANR = re.compile('<.*?>')
def cleanhtml(raw_html):
    cleantext = re.sub(CLEANR, '', raw_html)
    return cleantext
    
df["pt_answer"] = df["pt_answer"].apply(lambda x: cleanhtml(x))

df["pt_title"] = df["pt_title"].str.lower()
df["pt_answer"] = df["pt_answer"].str.lower()

df = df[['pt_title', 'pt_answer']]
df
```

```python

actor_training_data = []
rlhf_training_data = []
reward_training_data = []

for index, row in df.iterrows():
    actor = { "user_input": row['pt_title'],
              "completion": row['pt_answer'][0:2048] }
    
    rlhf = { "user_input": row['pt_title'] }
    
    reward = { "user_input": row['pt_title'],
               "completion": row['pt_answer'][0:2048],
               "score": 5.0 }
    
    actor_training_data.append(actor)
    
    rlhf_training_data.append(rlhf)
    
    reward_training_data.append(reward)
```

```python
import json

with open('actor_training_data.json', 'w') as f:
    json.dump(actor_training_data, f)

with open('rlhf_training_data.json', 'w') as f:
    json.dump(rlhf_training_data, f)

with open('reward_training_data.json', 'w') as f:
    json.dump(reward_training_data, f)
```
